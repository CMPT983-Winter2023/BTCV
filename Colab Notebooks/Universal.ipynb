```python
import nibabel as nib
import numpy as np
import torch
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def show_data(images, labels, methods, index):
    fig, axs = plt.subplots(len(methods)+1, len(images), figsize=(20,20))

    for i in range(len(images)):
        organ = images[i].split("/")[-1].split(".")[-3].split("_")[0]
        axs[0][i].set_title(organ, fontsize=16)
   
    for i in range(len(images)):
        organ = images[i].split("/")[-1].split(".")[-3].split("_")[0]
        
        for j in range(len(methods)+1):
            if j == 0:
                method = "ground_truth"
            else:
                method = methods[j-1] # subtract 1 to get the correct index from the methods list
                if method == "rand":
                    method = "random"
                    
            img = nib.load(images[i])
            img = np.array(img.get_fdata())
            img = torch.tensor(np.expand_dims(img, axis=0))
            label = nib.load(labels[images[i]][j]) # subtract 1 to get the correct index from the labels list
            label = np.array(label.get_fdata())
            label = torch.tensor(np.expand_dims(label, axis=0))
            img_shape = img.shape
            label_shape = label.shape
            cut1 = 0
            cut2 = 0
            unique_labels = torch.unique(label[0, :, :, :])
            for cut in range(label_shape[3]):
                if cut1 == 0 and len(torch.unique(label[0, :, :, cut])) == len(unique_labels):
                    cut1 = cut
                if cut1 != 0 and len(torch.unique(label[0, :, :, cut])) < len(unique_labels):
                    cut2 = cut - 1
            if cut2 != 0:
                cut = round((cut1 + cut2) / 2)
            else:
                cut = round((cut1 + cut) / 2)
            cmap = ListedColormap(['black', 'green']) # define a colormap for the labels
            mask = label[0, :, :, cut] > 0 # create a mask of the non-zero regions in the label tensor
            masked_label = np.zeros_like(label[0, :, :, cut], dtype=np.int32)
            masked_label[mask] = label[0, :, :, cut][mask] # apply the mask to the label tensor
            axs[j][i].imshow(img[0, :, :, cut].detach().cpu(), cmap="gray")
            axs[j][i].imshow(masked_label, cmap=cmap, alpha=0.5)
            axs[j][i].axis('off')
            if i == 0: # add method names to the left side of each row
                axs[j][i].text(-50, img_shape[1]//2, method, fontsize=14, ha='right', va='center')
    
    #for i in range(len(images)): # add organ names to the top of each column
    #    organ = images[i].split("/")[-1].split(".")[-3].split("_")[0]
    #    axs[0][i].text(img_shape[2]//2, img_shape[1]+50, organ, fontsize=14, ha='center', va='bottom')
    plt.savefig('segmentation_results'+str(index)+'.png')    
    plt.show()
```

```python
## test dict part
test_img = []
test_lbl = []
test_post_lbl = []
test_name = []
dataset_list = ['PAOT_10_inner']
data_txt_path = './dataset/dataset_list/'
data_root_path = './dataset/'
for item in dataset_list:
    for line in open(data_txt_path + item +'_test.txt'):
        name = line.strip().split()[1].split('.')[0]
        test_img.append(data_root_path + line.strip().split()[0])
        test_lbl.append(data_root_path + line.strip().split()[1])
        test_post_lbl.append(data_root_path + name.replace('label', 'post_label') + '.h5')
        test_name.append(name)
data_dicts_test = [{'image': image, 'label': label, 'post_label': post_label, 'name': name}
            for image, label, post_label, name in zip(test_img, test_lbl, test_post_lbl, test_name)]
print('test len {}'.format(len(data_dicts_test)))
```

```python
# out/rand/test_healthp_0/output/10_Decathlon/spleen_19/spleen_19_gt.nii.gz
# out/rand/test_healthp_0/output/10_Decathlon/spleen_19/spleen_19_result_v1.nii.gz

index = 2

methods = ["rand", "one_hot", "clip", "medclip", "clip_prompt", "clip_attention"]
images = [data['image'] for i, data in enumerate(data_dicts_test) if i%3==index]
labels = {}
for img in images:
  labels[img] = ["./out/"+method+"/test_healthp_0/output/10_Decathlon/"+img.split("/")[-1].split(".")[-3]+"/"+img.split("/")[-1].split(".")[-3]+"_result_v1.nii.gz" for method in methods]
  labels[img] = [img.replace("imagesTr","labelsTr")] + labels[img]
```

```python
show_data(images, labels, methods, index)
```
